```python
import random
import os

print('Winning Rules as follows:\n'
+ 'Rock vs paper-> paper wins\n'
+ 'Rock vs scissor-> Rock wins\n'
+ 'Paper vs scissor-> scissor wins.')

again = 'y'
Name = input('Username : ')
while again == 'y' :
    user_c = input('Enter your choise : ')
    user_c = user_c.lower()
    os.system('cls')

    if user_c != 'rock' and user_c != 'scissor' and user_c != 'paper':
        print('Please, enter a valid input\n'+'rock, scissor or paper')
    else :
        list_c = ['rock', 'scissor', 'paper']
        computer_c = random.choice(list_c).lower()
        print('Computer choise is : ' + computer_c)
        if user_c == 'rock' :
            if computer_c == 'rock' :
                print('Rock Vs Rock --> No winners')
            elif computer_c == 'scissor' :
                print('Rock Vs Scissor --> '+ Name + ' wins')
            elif computer_c == 'paper' :
                print('Rock Vs Paper --> Computer wins')

        elif user_c == 'paper' :
            if computer_c == 'paper' :
                print('Paper Vs Paper --> No winners')
            elif computer_c == 'rock' :
                print('Paper Vs Rock --> '+ Name + ' wins')
            elif computer_c == 'scissor' :
                print('Paper Vs Scissor --> Computer wins')

        elif user_c == 'scissor' :
            if computer_c == 'scissor' :
                print('Scissor Vs Scissor --> No winners')
            elif computer_c == 'paper' :
                print('Scissor Vs Paper --> '+ Name + ' wins')
            elif computer_c == 'rock' :
                print('Scissor Vs Rock --> Computer wins')

        input_user = input("Do you want to play again? (Y/N)")
        # if user input n or N then condition is True
        again = input_user.lower()
        if again == 'n' :
            break
print('Thanks for playing')

```
